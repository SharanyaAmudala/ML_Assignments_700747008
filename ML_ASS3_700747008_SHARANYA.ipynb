```python
#1.a.  creating random vector of size 15 having only Integers in the range 1-20. 
import numpy as np

random_vector = np.random.randint(1, 21, 15)
print('random vector of size 15 having only Integers in the range 1-20:')
print(random_vector)

#1. Reshaping the array to 3 by 5 
random_vector = random_vector.reshape(3,5)
#2. Printing array shape
print('\nreshaped array to 3 by 5')
print(random_vector)

#3. Replacing the max in each row by 0  
maxes = np.amax(random_vector, axis=1)
for i in range(3):
    random_vector[i][np.where(random_vector[i] == maxes[i])] = 0
print('\nArray after Replacing the max in each row by 0')
print(random_vector)

#4. Create a 2-dimensional array of size 4 x 3 (composed of 4-byte integer elements), also print the shape, type and data type of the array.
two_dimensional_array = np.arange(12).reshape(4, 3).astype(np.int32)

#Printing shape
print('\nShape of the two dimentional array:')
print(two_dimensional_array.shape)
#Printing type
print('\ntype of the two dimentional array:')
print(type(two_dimensional_array))
#Printing data type
print('\ndata type of the two dimentional array:')
print(two_dimensional_array.dtype)
```

```python
#1.b. program to compute the eigenvalues and right eigenvectors of a given square array.
a = np.array([[3, -2], [1, 0]])

eigenvalues, eigenvectors = np.linalg.eig(a)

print("Eigenvalues:")
print(eigenvalues)

print("\nRight Eigenvectors:")
print(eigenvectors)
```

```python
#1.c. Program to Compute the sum of the diagonal element of a given array.

arr = np.array([[0,1,2],[3,4,5]])

print('Sum of diagonal elements: ')
print(np.trace(arr))#trace will sum up the diagonal values.
```

```python
#1.d. program to create a new shape to an array without changing its data
arr = np.array([[1, 2], [3, 4], [5, 6]])


print("Reshaped array is:")
print(arr.reshape(2,3))#reshaping the array.
```

```python
#2. program to create a below chart of the popularity of programming Languages.
import matplotlib.pyplot as plt 

# Data to plot 
labels = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++' 
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
explode = (0.1, 0, 0, 0, 0, 0)  # explodes 1st slice 

# Plotting 
plt.pie(popularity, explode=explode, labels=labels,autopct='%1.1f%%', shadow=True, startangle=140,  
        wedgeprops = {"edgecolor" : "black",
                      'linewidth': 0.5,
                      'antialiased': True}) 
plt.show()#shows the pie chart
```
